```python
####### ローカルファイルをアップロードしてデータフレームに格納 ######
from google.colab import files
import pandas as pd

# ファイルのアップロード
uploaded = files.upload()

# アップロードされたファイル名を取得
uploaded_filename = list(uploaded.keys())[0]

# データフレームに読み込む (エンコーディングを指定)
df2 = pd.read_csv(uploaded_filename, encoding='Shift-JIS')

# データフレームの内容を表示
df2
```

```python
# df1とdf2をマージしてdfに格納
df = pd.merge(df1, df2, on='等級')
df
```

```python
# '所属部署名'ごとに平均年齢と平均月給を計算してdf3を作成
df3 = df.groupby('所属部署名').agg({'年齢': 'mean', '月給': 'mean'}).reset_index()


# '所属部署名'ごとに平均年齢と平均月給を計算してdf3を作成
df3 = df.groupby('所属部署名').agg({'年齢': 'mean', '月給': 'mean'}).reset_index()

# 平均年齢と平均月給を整数に四捨五入して変換
df3['平均年齢'] = df3['年齢'].round().astype(int)
df3['平均月給'] = df3['月給'].round().astype(int)

df3
```
